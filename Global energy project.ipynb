```python
import pandas as pd
```

```python
import matplotlib.pylab as plt
```

```python
energy = pd.read_csv('global_sustainable_energy.csv')
```

```python
energy
```

```python
print(energy.head())
```

```python
united_states = energy[(energy.Entity == 'United States')]
print(united_states)
```

```python
import matplotlib.pyplot as plt
import pandas as pd


plt.scatter(x=united_states['Year'], y=united_states['Electricity from renewables (TWh)'])
plt.xlabel('Year')
plt.ylabel('Renewable Energy')
plt.title('United States Renewable Energy Over Years')
plt.grid(True) 
plt.show()
```

```python
plt.scatter(x=united_states['Year'], y=united_states['Electricity from fossil fuels (TWh)'])
plt.xlabel('Year')
plt.ylabel('Fossil Fuels')
plt.title('United States Fuels Over Years')
plt.grid(True) 
plt.show()
```
